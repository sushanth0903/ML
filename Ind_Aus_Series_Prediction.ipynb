```python
Ind_master_df['Opposition'] = Ind_master_df['Opposition'].map(lambda x: x[2:])
```

```python
Ind_master_df.loc[:,'FormIndicator']  = Ind_master_df['Year'].map(lambda x:'High' if x >= 2018 else ('Medium'if x >= 2015 else 'Low'))
```

```python
testCases.loc[:,'FormIndicator']  = testCases['Year'].map(lambda x:'High' if x >= 2018 else ('Medium'if x >= 2015 else 'Low'))
```

```python
Ind_master_df.drop(['Team','Year'],axis=1,inplace=True)
testCases.drop(['Year'],axis=1,inplace=True)
Ind_master_df = Ind_master_df.append([testCases])
Ind_master_df.reset_index()
```

```python

Ind_master_df_Opposition = pd.get_dummies(Ind_master_df['Opposition'],prefix="Opposition")
#Adding the results to the master dataframe
Ind_master_df = pd.concat([Ind_master_df,Ind_master_df_Opposition],axis=1)
Ind_master_df.drop(['Opposition'],axis=1,inplace=True)

Ind_master_df_Result = pd.get_dummies(Ind_master_df['Result'],prefix="Result")
#Adding the results to the master dataframe
Ind_master_df = pd.concat([Ind_master_df,Ind_master_df_Result],axis=1)
Ind_master_df.drop(['Result'],axis=1,inplace=True)

Ind_master_df_Ground = pd.get_dummies(Ind_master_df['Ground'],prefix="Ground")
#Adding the results to the master dataframe
Ind_master_df = pd.concat([Ind_master_df,Ind_master_df_Ground],axis=1)
Ind_master_df.drop(['Ground'],axis=1,inplace=True)

Ind_master_df_Bat = pd.get_dummies(Ind_master_df['Bat'],prefix="Bat")
#Adding the results to the master dataframe
Ind_master_df = pd.concat([Ind_master_df,Ind_master_df_Bat],axis=1)
Ind_master_df.drop(['Bat'],axis=1,inplace=True)

Ind_master_df_FormIndicator = pd.get_dummies(Ind_master_df['FormIndicator'],prefix="FormIndicator")
#Adding the results to the master dataframe
Ind_master_df = pd.concat([Ind_master_df,Ind_master_df_FormIndicator],axis=1)
Ind_master_df.drop(['FormIndicator'],axis=1,inplace=True)

Ind_master_df_Toss = pd.get_dummies(Ind_master_df['Toss'],prefix="Toss")
#Adding the results to the master dataframe
Ind_master_df = pd.concat([Ind_master_df,Ind_master_df_Toss],axis=1)
Ind_master_df.drop(['Toss'],axis=1,inplace=True)

```

```python
# Importing test_train_split from sklearn library
from sklearn.model_selection import train_test_split
```

```python
# Putting feature variable to X
X_train = Ind_master_df[0:112].drop(['Result_won','Result_lost','Result_tied'],axis=1)

# Putting response variable to y
y_train = Ind_master_df[0:112]['Result_won']

X_test = Ind_master_df[112:].drop(['Result_won','Result_lost','Result_tied'],axis=1)

# Putting response variable to y
y_test = Ind_master_df[112:]['Result_won']

print(y_test)
# Splitting the data into train and test
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
# Importing random forest classifier from sklearn library
from sklearn.ensemble import RandomForestClassifier

# Running the random forest with default parameters.
rfc = RandomForestClassifier()
# fit
rfc.fit(X_train,y_train)
```

```python
# Making predictions
predictions = rfc.predict(X_test)
India_Win_Percentage = 100*(sum(predictions)/20)
print(India_Win_Percentage)
```
